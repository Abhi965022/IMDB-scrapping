```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = ('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

```python
r = requests.get(url)
```

```python
content = BeautifulSoup(r.content, 'html.parser')
```

```python
Movie_name = []
Time = []
Year = []
Metascore = []
Votes = []
Gross = []
Rating = []
Genre = []
```

```python
movie_data = content.find_all('div', attrs = {'class':'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.find('h3').a.text
    Movie_name.append(name)
    
    year = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    Year.append(year)
    
    time = store.find('span', class_ = 'runtime').text.replace('min', '').replace(' ','')
    Time.append(time)
    
    metascore = store.find('span', class_ = 'metascore').text.replace(' ','') if store.find('span', class_ = 'metascore') else '&&&&&&&&'
    Metascore.append(metascore)
    
    rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n','')
    Rating.append(rate)
    
    value = store.find_all('span', attrs = {'name':'nv'})
    vote = value[0].text
    Votes.append(vote)
    
    gross = value[1].text if len(value) >1 else '&&&&&&'
    Gross.append(gross)
    
    genre = store.find('span', class_ = 'genre').text.replace('\n','').replace(' ','')
    Genre.append(genre)
    
```

```python


data = pd.DataFrame({'Movie name':Movie_name, 'Year':Year, 'Time':Time, 'Rating':Rating, 'Genre':Genre, 'Metascore':Metascore, 'Votes':Votes, 'Gross':Gross})
```
